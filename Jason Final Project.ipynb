```python
import pandas as pd
import matplotlib.pyplot as plt


parks = pd.read_csv('parks.csv')


parks.columns = parks.columns.str.strip()


neighborhood_col = None
for col in parks.columns:
    if "neighborhood" in col.lower():
        neighborhood_col = col
        break

park_counts = parks.groupby(neighborhood_col).size().reset_index(name="park_count")


park_counts = park_counts.sort_values("park_count", ascending=False)

top10 = park_counts.head(10)


plt.figure(figsize=(9,6))
plt.barh(top10[neighborhood_col], top10["park_count"], color="green")
plt.xlabel("Number of Parks")
plt.ylabel("Neighborhoods")
plt.title("Top 10 Healthy Pittsburgh Neighborhoods by Park Count")
plt.gca().invert_yaxis()
plt.show()
```
