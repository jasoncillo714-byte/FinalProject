```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("AIR DATA.csv")
site_avg = df.groupby('site')['index_value'].mean().sort_values(ascending=False)
site_avg = site_avg.rename(index={'Flag Plaza': 'Flag Plaza (Central Business District)'})

# Taller figure
plt.figure(figsize=(10,8))
site_avg.plot(kind='bar', color='teal')
plt.title("Average Air Quality Index by Site")
plt.xlabel("Monitoring Site")
plt.ylabel("Average Index Value")
plt.tight_layout()
plt.show()
```
